```python
import pandas as pd
import numpy as np
from pycaret.classification import *
import mlflow

data = pd.read_csv("data/creditcard.csv")
data_unseen = data.sample(frac = 0.1, random_state = 42)
data = data.drop(data_unseen.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print("Train + Val -", data.shape, "\nTest -", data_unseen.shape)
```

```python
data.head()
```

```python
  expt_basic = setup(
      data = data, 
      target = 'Class',
      session_id=42,
      train_size=0.8,
      numeric_imputation="median",
      categorical_imputation="mode",
      fix_imbalance=True
  )
```

```python
models()
```

```python
best_classifier = compare_models(sort="F1", fold=5, include = ["lightgbm", "dt"], n_select = 2)
```

```python
tuned_models = []
for model in best_classifier:
    print(model)
    print()
    tuned = tune_model(model, fold = 5, optimize = "F1")
    tuned_models.append(tuned)
```

```python
lgbm = finalize_model(tuned_models[0])
dt = finalize_model(tuned_models[1])
```

```python
predict_model(lgbm);
```

```python
predict_model(dt);
```

```python
from pycaret.utils import check_metric

print("Light Gradient Boosting Machine")
pred_unseen_tuned = predict_model(lgbm, data = data_unseen)
print("Accuracy\t", check_metric(pred_unseen_tuned.Class, pred_unseen_tuned.Label, 'Accuracy'))
print("F1\t\t", check_metric(pred_unseen_tuned.Class, pred_unseen_tuned.Label, 'F1'))

print("Decision Tree")
pred_unseen_tuned = predict_model(dt, data = data_unseen)
print("Accuracy\t", check_metric(pred_unseen_tuned.Class, pred_unseen_tuned.Label, 'Accuracy'))
print("F1\t\t", check_metric(pred_unseen_tuned.Class, pred_unseen_tuned.Label, 'F1'))
```
