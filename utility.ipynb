```python
import pandas as pd
import numpy as np
from pycaret.classification import *

data = pd.read_csv("data/creditcard.csv")
data_unseen = data.sample(frac = 0.1, random_state = 42)
data = data.drop(data_unseen.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print("Train + Val -", data.shape, "\nTest -", data_unseen.shape)
```

```python
expt_basic = setup(
    data = data, 
    target = 'Class',
    session_id=42,
    train_size=0.8,
    numeric_imputation="median",
    categorical _imputation="mode",
    fix_imbalance=True,
    log_experiment = True,
    log_plots = True,
    log_data = True,
    experiment_name = "First"
)
```

```python
models()
```

```python
best_classifier = compare_models(sort="F1", fold = 5, n_select = 5)
```
